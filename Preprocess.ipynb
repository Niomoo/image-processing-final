```python
import json
import os, shutil
```

```python
full_path = 'class_data/Train/'
```

```python
path_powder_uncover = full_path + 'powder_uncover/label/'
path_powder_uneven = full_path + 'powder_uneven/label/'
path_scratch = full_path + 'scratch/label/'
```

```python
img_powder_uncover = full_path + 'powder_uncover/image/'
img_powder_uneven = full_path + 'powder_uneven/image/'
img_scratch = full_path + 'scratch/image/'
```

```python
count = 0
for img in os.listdir(img_powder_uncover):
    source = os.path.join(img_powder_uncover, img)
    target = full_path + 'data/' + str(count) + '.png'
    shutil.copy(source, target)
    count += 1
for img in os.listdir(img_powder_uneven):
    source = os.path.join(img_powder_uneven, img)
    target = full_path + 'data/' + str(count) + '.png'
    shutil.copy(source, target)
    count += 1
for img in os.listdir(img_scratch):
    source = os.path.join(img_scratch, img)
    target = full_path + 'data/' + str(count) + '.png'
    shutil.copy(source, target)
    count += 1
```

```python
counter = 0
output_path = full_path + 'data/'
for json_file in os.listdir(path_powder_uncover):
    with open(path_powder_uncover + json_file, 'r') as file:
        data = json.load(file)
        name = json_file.split('.')[0]
        with open(output_path + str(counter) + '.txt', 'w') as output_file:
            for box in data['shapes']:
                if box['label'] == 'powder_uncover':
                    label = 0
                    w = 3384
                    h = 3330
                    x1 = box['points'][0][0]
                    y1 = box['points'][0][1]
                    x2 = box['points'][1][0]
                    y2 = box['points'][1][1]
                    x_ = (x1 + x2) / (2 * w)
                    y_ = (y1 + y2) / (2 * h)
                    w_ = (x2 - x1) / w
                    h_ = (y2 - y1) / h
                elif box['label'] == 'powder_uneven':
                    label = 1
                    w = 3384
                    h = 3330
                    x1 = box['points'][0][0]
                    y1 = box['points'][0][1]
                    x2 = box['points'][1][0]
                    y2 = box['points'][1][1]
                    x_ = (x1 + x2) / (2 * w)
                    y_ = (y1 + y2) / (2 * h)
                    w_ = (x2 - x1) / w
                    h_ = (y2 - y1) / h
                elif box['label'] == 'scratch':
                    label = 2
                    w = 1254
                    h = 1244
                    x1 = box['points'][0][0]
                    y1 = box['points'][0][1]
                    x2 = box['points'][1][0]
                    y2 = box['points'][1][1]
                    x_ = (x1 + x2) / (2 * w)
                    y_ = (y1 + y2) / (2 * h)
                    w_ = (x2 - x1) / w
                    h_ = (y2 - y1) / h
                txt = str(label) + ' ' + str(x_) + ' ' + str(y_) + ' ' + str(w_) + ' ' + str(h_)
                output_file.write(txt + '\n')
        counter += 1
            
```
